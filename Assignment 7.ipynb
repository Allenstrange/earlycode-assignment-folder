# DEVELOP AN IQ TEST PROGRAM

```python
class IQ_test:
    def __init__ (self, Name, Age):
        self.Name = Name
        self.age = Age
        
    def quest(self):
        questions = {
        '(1) Which term will come in place of the question mark 3:5::7:?': {'A': '33', 'B': '15', 'C': '25', 'D': '31'},
        '(2) Find the missing term: poverty:Riches:Glory:?': {'A': 'Penury', 'B': 'Happiness', 'C': 'shame', 'D': 'Suffering'},
        '(3) Cowherd is to cattle as ___ is to sheep?': {'A': 'Shepherd', 'B': 'Watchman', 'C': 'Farmer', 'D': 'Dog'},
        '(4) Charles mother is Alena\'s father\'s sister. How is Charles mother related to Alena?': {'A': 'Paternal Aunt', 'B': 'Uncles Wife', 'C': 'Sister', 'D': 'Mother'},
        '(5) What number comes next in the sequence? 1-2-4-7-11-16-?': {'A': '22', 'B': '23', 'C': '24', 'D': '25'},
        '(6) At 5 o\'clock, what is the angle between the minute hand and the hour hand?': {'A': '150', 'B': '100', 'C': '120', 'D': '180'},
        '(7) Find the missing term: Temperature:Heat::Humidity:?' : {'A': 'Weather', 'B': 'Rain', 'C': 'Moisture', 'D': 'Night'},
        '(8) Find the missing term: French:France::Dutch:?' : {'A': 'Hungary', 'B': 'Sea', 'C': 'Holland', 'D': 'Hiroshima'},
        '(9) Finger is to hand as leaf is to ? ': {'A': 'Twig', 'B': 'Branch', 'C': 'Tree', 'D': 'Blossom'},
        '(10) What day was yesterday if monday is in two days ?': {'A': 'Wednesday', 'B': 'Thursday', 'C': 'Saturday', 'D': 'Friday'}
    
        }
        answers = ['D', 'C', 'A', 'A', 'C','A', 'C', 'C', 'A', 'D']
        score = 0
        k = -1
        total_score = 50
        for question, options in questions.items():
            print('\n',question, '\n','\n', options)
            answer = input('your answer: ')
            k += 1
            if answer == answers[k]:
                print('right')
                score += 5
            else:
                print('wrong')
                score += 0
        print()
        print(f'Got {score} out of {total_score} points')
        
    
        iq_score = (score/self.age) * 100
        if iq_score  <= 69:
            print('you are extremely low'.title())
        elif iq_score  >= 70 and iq_score <= 79:
            print('you are very low')
        elif iq_score  >= 80 and iq_score <= 89:
            print('you are Averagely low'.title())
        elif iq_score  >= 90 and iq_score <= 109:
            print('you are Average')
        elif iq_score  >= 110 and iq_score <= 119:
            print('you are Highly Average'.title())
        elif iq_score  >= 120 and iq_score <= 129:
            print('you are Very High'.title())
        else:
            print('you are extremely High'.title())
        
        
    
        
person1 = IQ_test(input('Name: '), int(input('Age: ')))
person1.quest()
     
        
        
        
```
